```python
import numpy as np
import matplotlib.pyplot as plt
from hist.intervals import poisson_interval
```

```python
size = 10**6
x = np.random.normal(size=size)
bins = np.linspace(-5, 5, 101)
x;
```

```python
n, bins, patches = plt.hist(x, histtype='step', bins=bins);
```

```python
binwidth = bins[1] - bins[0]
centers = bins[:-1] + 0.5*binwidth
plt.errorbar(centers, n, np.sqrt(n), );
lo, hi = poisson_interval(n, n)
plt.plot(centers, lo, linestyle='dashed', label='Poisson low')
plt.plot(centers, hi, linestyle='dashed', label='Poisson high')
plt.yscale('log')
plt.ylim(10**-4)
plt.legend();
```
