```python
import os
import matplotlib.pyplot as plt
from coffea.util import load
from pocket_coffea.utils.plot_utils import Shape
```

```python
filename = "/work/mmarcheg/BTVNanoCommissioning/output/pocket_coffea/pt_reweighting/pt_eta_tau21_reweighting_2018_withJES/output_all.coffea"
o = load(filename)
o.keys()
```

```python
year = "2018"
plot_dir = "plots_JES"
if not os.path.exists(plot_dir):
    os.makedirs(plot_dir)
for variable in ["pt", "eta", "tau21"]:
    histname = f"FatJetGoodNMuon1_{variable}"
    h_dict = o["variables"][histname]
    shape = Shape(h_dict, "tau21", histname, plot_dir)
    shape.build_stacks(year, "inclusive")
    shape.get_systematic_uncertainty()
    shape.define_figure()
    shape.plot_datamc(split_syst=True)
    outfile = os.path.join(plot_dir, f"{variable}_{year}.png")
    plt.savefig(outfile, dpi=300, format="png")
```
